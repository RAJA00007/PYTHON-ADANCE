scatter plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.style.use('dark_background')
```

```python
rollno = [1,2,3,4,5,6]
marks = [10,20,30,40,50,60]
```

```python
plt.figure(figsize=(12,8))
plt.scatter(rollno, marks, color='b', s=100, marker='d')
plt.show()
```

```python
temp_pune = [25,34,21,45,28,6,43,18,7,2]
humidity_pune = [28,25,29,20,26,50,19,29,52,55]
temp_bangalore = [34,35,36,37,28,27,26,25,31,20]
humidity_bangalore = [40,38,36,35,42,44,41,40,34,45]
```

```python
plt.figure(figsize=(8,8))
plt.xticks(np.arange(0,60,5))
plt.yticks(np.arange (10,60,5))
plt.scatter(temp_pune, humidity_pune, color = 'red', s=15)
plt.scatter(temp_bangalore, humidity_bangalore, color = 'yellow', s=15)
plt.xlabel('temp')
plt.ylabel('humidity')
plt.show()
```
