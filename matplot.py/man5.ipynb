```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
claases =[ 'maths', ' sci', 'hindi']
marks = [89,90,10]
```

```python
plt.pie(marks, labels=claases, colors =[ 'red','green','black'], autopct='%0.2f%%', explode=[0.002,0.01,0] , radius=2)
plt.show()
```
