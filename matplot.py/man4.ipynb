histogram

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
marks = np.random.randint(0,100,(50))
marks2 = np.random.randint(0,100,(50))
print(marks)
print(marks2)
```

```python
plt.hist(marks)
plt.show()
```

```python
bins = np.arange(0, 100, 5)
plt.figure(figsize=(6, 6))
plt.hist([marks, marks2], bins=bins, color=['orange','blue'], rwidth=0.8)
plt.xticks(np.arange(0, 100, 5))
plt.show()
```

```python

bins= np.arange(0,100,5)
plt.figure(figsize=(6,6))
plt.hist(marks,bins=bins, color ='orange', orientation='horizontal', histtype='step')
plt.yticks(np.arange(0,100,5))
plt.show()
```
