```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.style.use('dark_background')
```

```python
rollno = [1,2,3,4,5,6,7,8,9]
marks = [10,20,30,40,50,60,70,80,70]
plt.plot(rollno,marks, ':', linewidth=2)
plt.scatter(rollno,marks)
plt.show()
```

```python
rollno = [1,2,3,4,5,6,7,8,9]
marks = [10,20,30,40,50,60,70,80,90]
plt.plot(rollno,marks, ':', color = 'orange')
plt.show()
```
