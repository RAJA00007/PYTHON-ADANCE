scatter plot

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

df = sns.load_dataset('titanic')
df.head()
```

```python

plt.figure(figsize=(12,6))

sns.scatterplot(x='age', y = 'fare', data = df, hue = 'alive', style='class' ,alpha = 0.9)
sns.lineplot(x='age', y = 'fare', data = df, color = 'green')
plt.title('titanic data ananlysis')
```

BAR PLOT

```python
df.head(1)
```

```python
sns.barplot(x='class', y= 'fare', data = df, hue = 'sex', palette='inferno')
```

HEAT MAP

```python
flight_df = sns.load_dataset('flights')
flight_df.head()
     
```

```python
flight_df = flight_df.pivot(index="month", columns="year", values="passengers")
flight_df.head(12)

```

```python
plt.figure(figsize=(12,6))
ax = sns.heatmap(flight_df, annot=True, fmt='d', linecolor='k', linewidths='5', cmap='Reds', cbar=bool)
ax
```
