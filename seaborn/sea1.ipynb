```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
roll_no = [1,2,3,4,5,6,7,8,9]
marks = [25,45,67,89,0,9,23,12,33]
sampledf = pd.DataFrame({"Rollno":roll_no, "Marks":marks})
sampledf.head()
```

line plot

```python
sns.lineplot(x="Rollno", y="Marks", data = sampledf)
plt.title("student marks")
```
