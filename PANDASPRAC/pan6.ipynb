merge function

```python


# Sample DataFrames
df1 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40]
})

df2 = pd.DataFrame({
    'ID': [3, 4, 5, 6],
    'City': ['New York', 'Chicago', 'Los Angeles', 'Houston'],
    'Salary': [70000, 80000, 75000, 60000]
})

# Merge on common column 'ID'
merged_df = pd.merge(df1, df2, on='ID', how='inner')

# Display the result
print(merged_df)
```
